```python
plt.figure(figsize=(10, 10))
plt.plot(df.dteday, df['cnt'])
plt.xlabel('Date')
plt.ylabel('Number of Bike-Sharing Users in total')
plt.title('Bike-Sharing Users per Day')

plt.show()
```

```python
df['month'] = df['dteday'].dt.month
month_users = df.pivot_table(index='month', columns='dteday', values='cnt')
plt.figure(figsize=(10, 6))
sns.heatmap(month_users, cmap='YlOrRd')
plt.xlabel('Date')
plt.ylabel('Month')
plt.title('Bike-Sharing Users per Month')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['dteday'])
plt.figure(figsize=(10, 6))
plt.scatter(df['date'], df['cnt'])
plt.xlabel('Date')
plt.ylabel('Number of Users')
plt.title('Bike-Sharing Users per Day')
plt.show()
```

```python
holiday_df = df[df['holiday'] == 1]
non_holiday_df = df[df['holiday'] == 0]
```

```python
work=df[df['workingday']==1]
rest=df[df['workingday']==0]
plt.figure(figsize=(10, 6))
plt.plot(work['date'], work['cnt'], label='Holiday')
plt.plot(rest['date'], rest['cnt'], label='Non-Holiday')
plt.xlabel('Date')
plt.ylabel('Number of Users')
plt.title('Bike-Sharing Users per Day by Holiday')
plt.legend()
plt.show()
```
